This module was all about regression and using Python's scikitlearn library to build regression models.  Below, a dataset related to real estate prices in California is given. Your goal is to build a regression model to predict the price of a house in California.  After doing so, you are to *interpret* the model.  There are many strategies for doing so, including some built in methods from scikitlearn.  One example is `permutation_importance`. Take a look at the user guide for `permutation_importance` [here](https://scikit-learn.org/stable/modules/permutation_importance.html).  Use  the `sklearn.inspection` modules implementation of `permutation_importance` to investigate the importance of different features to your regression models. Remove lat and long from dataset and drop columns with NA values

```python
cali_clean = cali.drop(['longitude', 'latitude'], axis = 1).dropna()
cali_clean.columns
#sns.pairplot(cali_clean.select_dtypes(include = 'float64'), plot_kws=dict(alpha=.3))
```

Dummy/OHE code categorical variable

```python
onehot_encoder = OneHotEncoder(sparse_output = False)
dummy_variables = onehot_encoder.fit_transform(cali_clean[['ocean_proximity']])

cali_clean = cali_clean.drop('ocean_proximity', axis=1).reset_index().drop('index',axis=1)
cali_clean = pd.concat([cali_clean, pd.DataFrame(dummy_variables, columns = onehot_encoder.get_feature_names_out())], axis = 1)

print(cali_clean.columns)
cali_clean
```

Split into training, dev, test data

```python
from sklearn.utils import shuffle
num_rows = cali_clean.shape[0]
np.random.seed(100)
cali_train, cali_dev, cali_test = np.split(shuffle(cali_clean), [int(num_rows*.6), int(num_rows*.8)], axis=0)
```

Visualize dummy-other relationships (cali_train)

## Train regression model
Train regression model, with following:
- Drop longitude and latitude
- Use PCA to extract variance from 'total_rooms', 'population', and 'households'
- Keep all the one-hot features

### Standardize non-onehot features

```python
from sklearn.preprocessing import StandardScaler

cols_to_standardize = cali_train.columns[:7]
scaler = StandardScaler()
cali_train[cols_to_standardize] = scaler.fit_transform(cali_train[cols_to_standardize])

cali_train.mean()
cali_train.std() # good
```

### PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

cols_to_pca = cali_train.columns[1:5]

pca = PCA() # instantiate the PCA class
pca_column = pca.fit_transform(cali_train[cols_to_pca])
pca.explained_variance_ratio_ # good; most of the variance captured in the first column. call it PCA_1 and move on.
```

```python
pca_df = pd.DataFrame(pca_column[:,1], index = cali_train.index, columns = ['PCA_1'])
all(pca_df.index == cali_train.index) # good

cali_train = pd.concat([cali_train, pca_df], axis = 1)

sns.heatmap(cali_train.iloc[:,[1,2,3,4,6,12]].corr(method = 'spearman'), annot = True, center = 0, cmap = 'coolwarm')
```

### Train regression models
Try multiple polynomial models with cali_train, except drop total_rooms, total_bedrooms, population, and households

```python
model_columns = cali_train.columns[[0,5,7,8,9,10,11,12,6]]

# final heatmap of correlations b/t variables
sns.heatmap(cali_train[model_columns].corr(method='spearman'), annot = True, center = 0, cmap = 'coolwarm')
```

```python
# train regression models
from sklearn import linear_model

X_train = cali_train[model_columns].drop('median_house_value', axis=1)
y_train = cali_train[['median_house_value']]
```

#### Prep test dataset

```python
# standardize
cols_to_standardize = cali_dev.columns[:7]
scaler = StandardScaler()
cali_dev[cols_to_standardize] = scaler.fit_transform(cali_dev[cols_to_standardize])

print(cali_dev.std())
cali_dev.mean()
```

```python
# PCA
cols_to_pca = cali_dev.columns[1:5]

pca = PCA() # instantiate the PCA class
pca_column = pca.fit_transform(cali_dev[cols_to_pca])
pca.explained_variance_ratio_ # good; most of the variance captured in the first column again
```

```python
pca_df = pd.DataFrame(pca_column[:,1], index = cali_dev.index, columns = ['PCA_1'])
all(pca_df.index == cali_dev.index) # good

cali_dev = pd.concat([cali_dev, pca_df], axis = 1)

# X_dev and y_dev
model_columns = cali_dev.columns[[0,5,7,8,9,10,11,12,6]]

X_dev = cali_dev[model_columns].drop('median_house_value', axis=1)
y_dev = cali_dev[['median_house_value']]
```

Function for run loop for diff polynomial models

```python
from sklearn.metrics import mean_squared_error

def simple_cross_validation(X_train, y_train, X_test, y_test, complex_max: int = 5):
    best_pipe = None #placeholder for best model
    best_mse = np.inf #set best mse to infinity to begin
    mse_set = []
    complex_max += 1
    for i in range(1, (complex_max)):
        pipe = Pipeline([('pfeat', PolynomialFeatures(degree = i, include_bias=False)),
                         ('linreg', LinearRegression())])
        pipe.fit(X_train, y_train)
        test_preds = pipe.predict(X_test)
        test_mse = mean_squared_error(y_test, test_preds)
        if test_mse < best_mse:
            best_mse = test_mse
            best_pipe = pipe
        mse_set.append(test_mse)
        print(f"\rProgress: {i}/{complex_max-1}", end='')
    mse_set = pd.DataFrame(data = mse_set, index = list(range(1, complex_max)))
    return best_pipe, mse_set
```

```python
sample_valid, mse_set = simple_cross_validation(X_train, y_train, X_dev, y_dev, complex_max = 5)
sample_valid.get_params()
mse_set
```

```python
# degree of 1 had lowest errors; run regular multiple regression
train_model = linear_model.LinearRegression() # instantiate the model; same as with simple regression
train_model.fit(X_train, y_train) 

train_model.coef_ # feature coefficients
train_model.intercept_
train_model.score(X_dev, y_dev)

print(X_train.columns)
X_dev.columns
```

```python
# what about ridge regression?
from sklearn.linear_model import Ridge

train_ridge = Ridge(alpha=1e-2).fit(X_train, y_train)
train_ridge.score(X_dev, y_dev)
```

Ridge regression performed very slightly worse, after the 6th decimal place. Stick with simple linear regression.

### Calculate permutation importance scores

```python
from sklearn.inspection import permutation_importance

r = permutation_importance(train_model, X_dev, y_dev, n_repeats=30, random_state=0)

for i in r.importances_mean.argsort()[::-1]:
  #if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
    print(f"{X_dev.columns[i]:<8}: "
          f"{r.importances_mean[i]:.3f}"
          f" +/- {r.importances_std[i]:.3f}")
```
