- P: # of lagged observations of seasonal component. Consistenty yearly cycle; P =1
- D: Degree of seasonal diffrencing. Set at 1.
- Q: # of seasonal moving average terms. Not sure which is best; going with 1.

# GridSearch SARIMAX

```python
import itertools
import pandas as pd
import statsmodels.api as sm
from datetime import datetime
```

```python
# search for optimal seasonal_order values, between 0 - 3 each

# Define the range of parameters to search
p = d = q = range(0, 3)  # ranges for P, D, Q
s = [12]  # yearly seasonality

# Generate all different combinations of p, d, q and s
pdq_combinations = list(itertools.product(p, d, q, s))

lowest_aic = float('inf')
best_pdq = None
best_model_summary = None

gridsearch_start = datetime.now()
for combination in pdq_combinations:
    try:
        # The seasonal_order requires four parameters (P, D, Q, S)
        model = SARIMAX(X_train,
                        order=(0, 1, 1),
                        seasonal_order=combination)

        model_fit = model.fit(disp=False)  # disp=False to avoid verbose output

        # Check if the current model has the lowest AIC
        if model_fit.aic < lowest_aic:
            lowest_aic = model_fit.aic
            best_pdq = combination
            best_model_summary = model_fit.summary()

    except Exception as e:
        continue
gridsearch_end = datetime.now()
# Print the best model's parameters and summary
print(f"Time to perform grid search: {gridsearch_end - gridsearch_start}")
print(f'Best SARIMAX parameters: {best_pdq}')
print(f'Lowest AIC: {lowest_aic}')
print(best_model_summary)
```

```python
# make SARIMAX model with optimal parameters

sarimax2 = SARIMAX(X_train, order=(0, 1, 1), seasonal_order = (0, 2, 2, 12)).fit(disp=False)
sarimax2.summary()
```

```python
fig, ax = plt.subplots()
plt.plot(sarimax2.predict(), label='SARIMAX Prediction')
plt.plot(X_train, label='Time Series')
plt.legend()
ax.figure.autofmt_xdate()

plt.title('SARIMAX_2 Prediction vs. Actual')
plt.ylabel('# of employees')
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
# If using `mean_squared_error()`, need to make sure the series align first

sarimax2_pred_error = (sarimax2.predict() - X_train['Employees']).iloc[1:] # drop the first; prediction is 1
aligned_actual, aligned_predict = X_train['Employees'].align(sarimax2_pred_error, join='inner')
sarimax2_pred_RMSE = np.sqrt(mean_squared_error(aligned_actual, aligned_predict))
print(f"Prediction RMSE is {sarimax2_pred_RMSE}, whic is {sarimax2_pred_RMSE/X_train['Employees'].mean()}% of the mean.")
```

```python
sarimax2_forecast = sarimax2.forecast(steps=len(X_test))
fig, ax = plt.subplots()
plt.plot(sarimax2_forecast, label='SARIMAX_2 Forecast')
plt.plot(X_test, label='Time Series')
plt.legend()
ax.figure.autofmt_xdate()

plt.title('SARIMAX_2 Forecast vs. Actual')
plt.ylabel('# of employees')
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
#sarimax1_MSE = mean_squared_error(X_test['Employees'], sarimax1_forecast) # wrong

# If using `mean_squared_error()`, need to make sure the series align first
aligned_actual, aligned_predict = X_test['Employees'].align(sarimax2_forecast, join='inner')
sarimax2_RMSE = np.sqrt(mean_squared_error(aligned_actual, aligned_predict))
print(f"Forecast RMSE is {sarimax2_RMSE}, whic is {sarimax2_RMSE/X_test['Employees'].mean()}% of the mean.")
```

# Conclusion

The GridSearch method did not identify any parameters within the tested range that performed notably better than the seasonal parameters identified by the rational-empirical approach. The GridSearch method settled on higher-degree models with more complexity. Between the SARIMAX models, the difference in RMSE on the test range was less than 3-10ths of a unit (# of employees), and the RMSE of the forecast was better for the first SARIMAX model, not the model identified by the GridSearch.

Therefore, the first model was deemed to be better.
