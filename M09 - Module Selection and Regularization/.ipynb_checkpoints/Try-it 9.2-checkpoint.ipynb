Experience and age are highly correlated with each other. Identify importance with permutation importance and remove the less important one.

````python
from sklearn.utils import shuffle

temp_X = pd.DataFrame(StandardScaler().fit_transform(data[['AGE','EXPERIENCE']]), columns = ['AGE', 'EXPERIENCE'])
temp_data = pd.concat([temp_X, data['WAGE']], axis=1)

num_rows = temp_X.shape[0]
np.random.seed(42)
temp_train, temp_dev = np.split(shuffle(temp_data), [int(num_rows*.7)], axis=0)

temp_X_train = temp_train[['AGE', 'EXPERIENCE']]
temp_y_train = temp_train['WAGE']

temp_X_dev = temp_dev[['AGE', 'EXPERIENCE']]
temp_y_dev = temp_dev['WAGE']

temp_X_train.head()
````

````python
temp_model = LinearRegression()
temp_model.fit(temp_X_train, temp_y_train)
r = permutation_importance(temp_model, temp_X_dev, temp_y_dev, n_repeats=30, random_state=0)
for i in r.importances_mean.argsort()[::-1]:
  if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
    print(f"{temp_X_train.columns[i]:<8}: "
          f"{r.importances_mean[i]:.3f}"
          f" +/- {r.importances_std[i]:.3f}")
````

AGE has higher importance, which makes sense since it has a slightly higher bivariate correlation with WAGE.

Keep AGE, drop EXPERIENCE, and continue. Below: prep training and dev data to use

````python
std_columns = ['EDUCATION', 'EXPERIENCE', 'AGE']
data_std = pd.DataFrame(StandardScaler().fit_transform(data[std_columns]), columns = std_columns)
data_std = pd.concat([data_std, data[categorical_features + ['WAGE']]], axis=1)

num_rows = data_std.shape[0]
np.random.seed(42)
data_train, data_dev = np.split(shuffle(data_std), [int(num_rows*.7)], axis=0)

X_train, y_train, X_dev, y_dev = data_train.drop('WAGE', axis=1), data_train['WAGE'], data_dev.drop('WAGE', axis=1), data_dev['WAGE']

print(X_train.columns == X_dev.columns)
print(y_train[:5])
X_train.head()
````

# Ridge regression: Drop EXPERIENCE

````python
from sklearn.compose import ColumnTransformer

no_exp_pipe = Pipeline([
    ('transformer', ColumnTransformer(remainder='passthrough', transformers=[('cat', OneHotEncoder(), ['SOUTH', 'SEX', 'UNION', 'RACE', 'OCCUPATION', 'SECTOR', 'MARR'])])),
    ('model', TransformedTargetRegressor(func=np.log, inverse_func=np.exp, regressor=Ridge()))
])
param_grid = {'model__regressor__alpha': 10**np.linspace(-5, 4, 10)}

model_finder = GridSearchCV(estimator = no_exp_pipe,
                            param_grid = param_grid,
                            scoring = 'neg_mean_squared_error',
                            error_score='raise',
                            cv = 10)
model_finder.fit(X_train.drop('EXPERIENCE', axis=1), y_train)
````

# With Experience

````python
exp_pipe = Pipeline([
    ('transformer', ColumnTransformer(remainder='passthrough', transformers=[('cat', OneHotEncoder(), ['SOUTH', 'SEX', 'UNION', 'RACE', 'OCCUPATION', 'SECTOR', 'MARR'])])),
    ('model', TransformedTargetRegressor(func=np.log, inverse_func=np.exp, regressor=Ridge()))
])
param_grid = {'model__regressor__alpha': 10**np.linspace(-5, 4, 10)}

exp_model_finder = GridSearchCV(estimator = exp_pipe,
                                param_grid = param_grid,
                                scoring = 'neg_mean_squared_error',
                                error_score='raise',
                                cv = 10)
exp_model_finder.fit(X_train, y_train)
````

````python
exp_best_est = exp_model_finder.best_estimator_
exp_best_est
````

````python
exp_regressor = exp_best_est.named_steps['model'].regressor_
exp_regressor
````

````python
exp_coefs = exp_regressor.coef_
exp_features = exp_best_est.named_steps['transformer'].get_feature_names_out()
exp_coef_df = pd.DataFrame(exp_coefs, index = exp_features, columns = ['coefficient'])
exp_coef_df
````

````python
exp_train_mse = mean_squared_error(exp_model_finder.best_estimator_.predict(X_train), y_train)
exp_dev_mse = mean_squared_error(exp_model_finder.best_estimator_.predict(X_dev), y_dev)
exp_dev_r2 = r2_score(y_dev, exp_model_finder.best_estimator_.predict(X_dev))
print(f"Exp Training MSE: {exp_train_mse}")
print(f"Exp Dev MSE: {exp_dev_mse}")
print(f"Exp Dev R^2: {exp_dev_r2}")
print(exp_regressor.get_params()['alpha'])
````
